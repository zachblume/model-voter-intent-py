```python
# TEST SIZE PERCENTAGE
test_size_percentage = .3

# Load file and split into test and train dataframes
import pandas as pd
from sklearn.model_selection import train_test_split

url='/content/Training Data May 3 all ids for machine learning project Rana - fullestexport3apr1520220501-706310213 (1).csv'
df = pd.read_csv(url)
target_name='Support_Rana?'
y = df[target_name]
X = df.drop(columns=[target_name])
#X=X.fillna('UNK')
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = test_size_percentage,random_state = 123)

# Isolate column datatypes
from sklearn.compose import make_column_selector as selector

numerical_columns_selector = selector(dtype_exclude=object)
categorical_columns_selector = selector(dtype_include=object)

numerical_columns = numerical_columns_selector(X)
categorical_columns = categorical_columns_selector(X)

from sklearn.preprocessing import OrdinalEncoder, StandardScaler

categorical_preprocessor = OrdinalEncoder()
numerical_preprocessor = StandardScaler()



import numpy as np
import sklearn.datasets
import sklearn.compose
import sklearn.tree
import sklearn.impute

numeric_imputer = sklearn.impute.SimpleImputer(strategy='mean')
numeric_scaler = sklearn.preprocessing.StandardScaler()

nominal_imputer = sklearn.impute.SimpleImputer(strategy='constant', fill_value='missing')
nominal_encoder = sklearn.preprocessing.OrdinalEncoder()

numeric_transformer = sklearn.pipeline.make_pipeline(numeric_imputer, numeric_imputer)
nominal_transformer = sklearn.pipeline.make_pipeline(nominal_imputer, nominal_encoder)

from sklearn.compose import ColumnTransformer

from sklearn.compose import make_column_transformer
preprocessor = make_column_transformer(
     (numeric_transformer, numerical_columns),
     (nominal_transformer, categorical_columns))
     

from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import make_pipeline

# Generate all models

#from lazypredict.Supervised import LazyClassifier
#clf = make_pipeline(preprocessor, LazyClassifier())

#clf = LazyClassifier()


import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
from sklearn.metrics import classification_report

#models = clf.fit(X_test,y_test)
#models,predictions = clf.fit(X_train, X_test, y_train, y_test)
#model_dictionary = clf.provide_models(X_train,X_test,y_train,y_test)
#models




# Select the best model

# Generate confusion diagram

# Export predictions to csv

```
